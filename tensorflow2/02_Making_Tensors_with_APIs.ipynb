```python
import tensorflow as tf
import numpy as np
```

```python
test_list = [1, 1, 1, 1, 1, 1]

t1 = tf.constant(test_list)
print(t1)
```

```python
t2 = tf.ones(shape=(100,))
print(t2)
```

```python
t2 = tf.ones(shape=(128, 128, 3))
print(t2)
```

```python
t4 = tf.ones(shape=(128, 128, 3)) * 3
print(t4)

"""
scala 곱하기 vector(tensor)는 각 element마다 전부 scala가 곱해진다.
"""
```

```python
test_list = [[1, 2, 3], [4, 5, 6]]

t1 = tf.Variable(test_list)
print(t1)

t2 = tf.ones(shape=t1.shape)
print(t2)
```

```python
t2 = tf.ones_like(t1)
print(t2)
```

```python
t3 = tf.zeros_like(t1)
print(t3)
```

```python
t1 = tf.random.normal(shape=(10, 10))
print(t1)
```

```python
# np.random.seed(1) # numpy에서의 random 출력값 고정
tf.random.set_seed(1) # tensorflow에서의 random 출력값 고정
```

```python
tf.random.set_seed(1)

t1 = tf.random.normal(shape=(3, 3))
print(t1)

tf.random.set_seed(1)

t1 = tf.random.normal(shape=(3, 3))
print(t1)

"""같은 결과가 나옴"""
```

```python
tf.random.set_seed(1)

t1 = tf.random.normal(shape=(3, 3))
print(t1)

t1 = tf.random.normal(shape=(3, 3))
print(t1)

"""다른 결과가 나옴"""
```

```python
t2 = tf.random.normal(mean=3, stddev=1, shape=(1000,))
print(t2)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 10))
ax.hist(t2.numpy(), bins=30)

ax.tick_params(labelsize=10)
```

```python
t2 = tf.random.uniform(shape=(10000,), minval=-10, maxval=10)

fig, ax = plt.subplots(figsize=(10, 10))
ax.hist(t2.numpy(), bins=30)

ax.tick_params(labelsize=10)
```

```python
t2 = tf.random.poisson(shape=(1000,), lam=5)

fig, ax = plt.subplots(figsize=(10, 10))
ax.hist(t2.numpy(), bins=30)

ax.tick_params(labelsize=10)
```

```python
t1 = tf.random.normal(shape=(128, 128, 3))

print('t1.shape :', t1.shape)
print('t1.dtype :', t1.dtype)
```

```python
test_np = np.random.randint(-10, 10, size=(100,))
print(test_np.dtype)

t1 = tf.constant(test_np)
print(t1.dtype)

"""
tensorflow에서는 기본적으로 input data의 dtype이 float32가 들어오길 원한다.
따라서, int형인 경우에는 tensorflow에서 learning이 잘 되지 않을 수 있다.
"""
```

```python
test_np = np.random.randint(-10, 10, size=(100,))
print(test_np.dtype)

t1 = tf.constant(test_np, dtype=tf.float32)
print(t1.dtype)

"""
위의 dtype 이슈를 방지하기 위해서 변환시에 dtype을 지정해줄 수 있다.
"""
```
