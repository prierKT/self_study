```python
import tensorflow as tf
import numpy as np
from tensorflow.python.ops.gen_math_ops import mean
```

```python
t1 = tf.constant([1, 2, 3])
t2 = tf.constant([10, 20, 30])

print(t1 + t2)
```

```python
lst1 = [1, 2, 3]
lst2 = [10, 20, 30]

print(lst1 + lst2)

"""
list끼리 더하면 concatenation으로 만들어 진다.
tensor끼리 더하면 각 element를 연산한다.
"""
```

```python
print(t1 + t2)
print(t1 - t2)
print(t1 * t2)
print(t1 / t2)
print(t1 % t2) # 나머지
print(t1 // t2) # 몫
```

```python
# shape이 완전히 같은 경우
t1 = tf.random.normal(shape=(3, 4), mean=0, stddev=5)
t2 = tf.random.normal(shape=(3, 4), mean=0, stddev=5)

t1 = tf.cast(t1, dtype=tf.int16)
t2 = tf.cast(t2, dtype=tf.int16)

print(t1.numpy())
print(t2.numpy())

print(t1 + t2)
```

```python
# columns의 수가 같은 경우
t1 = tf.random.normal(shape=(3, 4), mean=0, stddev=5)
t2 = tf.random.normal(shape=(1, 4), mean=0, stddev=5)

t1 = tf.cast(t1, dtype=tf.int16)
t2 = tf.cast(t2, dtype=tf.int16)
t3 = t1 + t2

print(t1.numpy(), "\n")
print(t2.numpy(), "\n")
print(t3.numpy(), "\n")
```

```python
# rows의 수가 같은 경우
t1 = tf.random.normal(shape=(3, 4), mean=0, stddev=5)
t2 = tf.random.normal(shape=(3, 1), mean=0, stddev=5)

t1 = tf.cast(t1, dtype=tf.int16)
t2 = tf.cast(t2, dtype=tf.int16)
t3 = t1 + t2

print(t1.numpy(), "\n")
print(t2.numpy(), "\n")
print(t3.numpy(), "\n")
```

```python
# rows와 columns의 수가 같은 경우
t1 = tf.random.normal(shape=(3, 1), mean=0, stddev=5)
t2 = tf.random.normal(shape=(1, 4), mean=0, stddev=5)

t1 = tf.cast(t1, dtype=tf.int16)
t2 = tf.cast(t2, dtype=tf.int16)
t3 = t1 + t2

print(t1.numpy(), "\n")
print(t2.numpy(), "\n")
print(t3.numpy(), "\n")

"""
tensorflow에서 간단한 tensor는 numpy나 numerical computation library처럼,
element-wise가 기본적인 operation으로 setting 되어 있다.
그리고 broadcasting은 연산을 행하려는 두 행렬의 shape이 맞지 않더라도
한 쪽을 다른 한 쪽에 맞출 수 있는 상황이라면 차원을 확장시켜 연산이 가능하도록 할 수 있다.
"""
```

```python
t1 = tf.random.normal(shape=(3, 4), mean=0, stddev=5)
t1 = tf.cast(t1, dtype=tf.int16)

print(t1)
```

```python
tf.math.reduce_sum
tf.math.reduce_prod

tf.math.reduce_max
tf.math.reduce_min

tf.math.reduce_mean
tf.math.reduce_std
tf.math.reduce_variance

tf.math.reduce_all
tf.math.reduce_any
```

```python
t1 = tf.random.normal(shape=(3, 4), mean=0, stddev=5)
t1 = tf.cast(t1, dtype=tf.int16)

t2 = tf.math.reduce_sum(t1) # 모든 elements의 합
t3 = tf.math.reduce_sum(t1, axis=0) # row를 없애라
t4 = tf.math.reduce_sum(t1, axis=1) # column을 없애라

print(t1.numpy(), t1.shape, "\n")
print(t2.numpy(), t2.shape, "\n")
print(t3.numpy(), t3.shape, "\n")
print(t4.numpy(), t4.shape, "\n")
```

```python
t1 = tf.random.normal(shape=(128, 128, 3), mean=0, stddev=5)
t1 = tf.cast(t1, dtype=tf.int16)

t2 = tf.math.reduce_sum(t1, axis=2) # 3차원 일때, 채널을 없애라

print(t1.shape, "\n")
print(t2.shape, "\n")
```
