```python
import tensorflow as tf

from tensorflow.keras.models import Model, Sequential

from tensorflow.keras.layers import Layer, Dense, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Activation

from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import SparseCategoricalAccuracy
```

```python
class CustomLayer1(Layer):
  def __init__(self):
    super(CustomLayer1, self).__init__()
    
    self.dense = Dense(units=128, activation='relu')
    
  def call(self, x):
    predictions = self.dense(x)
    return predictions
```

복잡한 모델을 설계할 때, 동일한 레이어들이 반복되는 구간을 하나의 모듈로 만들어 사용한다면 전체적인 코드를 훨씬 가시성이 좋게 만들 수 있고, 유지보수 측면에서도 더 효율적으로 관리할 수 있다. 이렇게 만들어진 커스템 레이어를 Sequential, Fucntional, Model Sub-classing과 조합하여 다양한 방식으로 효율적인 모델을 빌드할 수 있다.

```python
model = Sequential()
model.add(CustomLayer1())
```

Sub-classing을 사용할 때, tf.keras.models.Model을 상속받는다면 fit(), evaluate(), predict() 메소드를 사용할 수 있는 반면, tf.keras.layers.Layer를 상속받는 경우에는 이 기능을 사용할 수 없다.
