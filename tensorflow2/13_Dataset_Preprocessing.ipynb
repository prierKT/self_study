```python
a = [1, 2, 3, 4, 5]

def double(in_val):
  return 2 * in_val

# map
result1 = list(map(double, a))
print(result1)

# lambda
result2 = list(map(lambda x: 2 * x, a))
print(result2)

# tf.cast
t1 = tf.constant([1, 2, 3, 4, 5])
print(t1.dtype)

t2 = tf.cast(t1, tf.float32)
print(t2.dtype)
```

```python
def mnist_data_loader():
  
  def normaliaztion(images, labels):
    images = tf.cast(images, tf.float32) / 255.
    return [images, labels]
  
  train_ds, test_ds = tfds.load(
                                name='mnist',
                                shuffle_files=True,
                                as_supervised=True,
                                split=['train', 'test'],
                                batch_size=8)
  
  train_ds = train_ds.map(normaliaztion)
  test_ds = test_ds.map(normaliaztion)
  
  return train_ds, test_ds
```

```python
train_ds, test_ds = mnist_data_loader()

for images, labels in train_ds:
  print(images.shape)
  print(images.dtype)
  break
```
