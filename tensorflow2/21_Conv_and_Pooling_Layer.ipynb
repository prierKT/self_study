```python
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D
```

```python
conv = Conv2D(filters=8, kernel_size=3, padding='valid', activation='relu')
pool = MaxPooling2D(pool_size=2, strides=2)
```

```python
image = tf.random.normal(mean=0, stddev=1, shape=(1, 28, 28, 3))
print(image.shape)
```

```python
conved = conv(image)
print(conved.shape)
```

```python
print('w :',conv.get_weights()[0].shape) # weights
print('b :',conv.get_weights()[1].shape) # bias
```

```python
pooled = pool(conved)
print(pooled.shape)
```

```python
conv1 = Conv2D(filters=8, kernel_size=3, padding='valid', activation='relu')
pool1 = MaxPooling2D(pool_size=2, strides=2)

conv2 = Conv2D(filters=16, kernel_size=3, padding='valid', activation='relu')
pool2 = MaxPooling2D(pool_size=2, strides=2)

conv3 = Conv2D(filters=32, kernel_size=3, padding='valid', activation='relu')
pool3 = MaxPooling2D(pool_size=2, strides=2)
```

```python
print()
print(image.shape)
conved = conv1(image)
print(conved.shape)
pooled = pool1(conved)
print(pooled.shape)
```

```python
conved = conv2(pooled)
print(conved.shape)
pooled = pool2(conved)
print(pooled.shape)
```

```python
conved = conv3(pooled)
print(conved.shape)
pooled = pool3(conved)
print(pooled.shape)
```

```python
model = Sequential()
model.add(Conv2D(filters=8, kernel_size=3, padding='valid', activation='relu'))
model.add(MaxPooling2D(pool_size=2, strides=2))

model.build(input_shape=(None, 28, 28, 1))
model.summary()
```
